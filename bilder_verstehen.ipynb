```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import imageio.v2 as imageio
```

# RGB Bilder

```python
img = imageio.imread('images/pencils.png')
plt.imshow(img)
plt.xticks([])
plt.yticks([])
```

```python
crop_R = np.copy(img)
crop_R[:, :, 1] = 0
crop_R[:, :, 2] = 0

plt.imshow(crop_R)
plt.xticks([])
plt.yticks([])
```

```python
crop_G = np.copy(img)
crop_G[:, :, 0] = 0
crop_G[:, :, 2] = 0

plt.imshow(crop_G)
plt.xticks([])
plt.yticks([])
```

```python
crop_B = np.copy(img)
crop_B[:, :, 0] = 0
crop_B[:, :, 1] = 0

plt.imshow(crop_B)
plt.xticks([])
plt.yticks([])
```

```python
crop_R_float = crop_R.astype(np.float32)/255
crop_G_float = crop_G.astype(np.float32)/255
crop_B_float = crop_B.astype(np.float32)/255

new_crop = 1.0 * crop_R_float + 1.0 * crop_G_float + 0.4 * crop_B_float

plt.imshow(new_crop)
plt.xticks([])
plt.yticks([])
print(new_crop.shape)
```

# Bilder Zusammenfügen

```python
img_corgy = imageio.imread('images/corgy.png').astype(np.float32)/255
img_pencils = imageio.imread('images/pencils.png').astype(np.float32)/255

new_image = 0.25 * img_pencils + 0.75 * img_corgy

plt.imshow(new_image)
plt.xticks([])
plt.yticks([])
plt.show()
```
